```python
test = pd.DataFrame([[1,2,3], [4,5,6], [7,8,8]], columns = ['A', 'A', 'C'], index = ['D', 'D', 'F'])
test.reset_index()
```

```python
def transform(tb, key):
    nrg = pd.melt(test.reset_index(), id_vars = ['index'])
    dst = pd.melt(test.reset_index(), id_vars = ['index'])
    
    colname = '_'.join([key[1][0], key[1][1]])
    try:
        tb.insert(tb.shape[1], 'len_tcr_region', '')
        tb.insert(tb.shape[1], 'pos_tcr_region', '')
        tb.insert(tb.shape[1],  'aa_tcr_region', '')
        tb.insert(tb.shape[1], 'len_' + colname, '')
        tb.insert(tb.shape[1], 'pos_' + colname, '')
        tb.insert(tb.shape[1],  'aa_' + colname, '')
        tb.insert(tb.shape[1], 'nrg_' + colname, '')
        tb.insert(tb.shape[1], 'dst_' + colname, '')
    except ValueError:
        print 'COLUMN ALREADY EXISTS'
    
    cond = (tb['tcr_chain_name'] == key[0][0]) & (tb['tcr_region'] == key[0][1])
    cut_index = tb[cond].index.tolist()[0]
    #if nor any(tb[cond][])
    res = pd.concat([tb.iloc[:cut_index, :], pd.concat(nrg.shape[0]*[tb[cond]]), tb.iloc[cut_index + 1:, :]])
    res.loc[cond, 'len_tcr_region'] = 2
    
    return res
```

```python
transform(table, (('alpha', 'CDR3'), ('alpha', 'CDR1')))
```
